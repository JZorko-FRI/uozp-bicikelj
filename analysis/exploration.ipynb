```python
def construct_query(values, columns):
    """
    Construct query for selecting data from a split.
    """
    # Construct query for selecting data from a split
    conditions = []
    for value, column in zip(values, columns):
        conditions.append(f'`{column}` == {repr(value)}')
    return ' & '.join(conditions)
```
